```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./data/adm_data.csv')
```

```python
df.columns
```

```python
df.head(10)
```

```python
sns.lmplot(x='GRE Score', y='CGPA',data=df)
```

We color the data points to do the applicants segmentation:

```python
sns.lmplot(x='GRE Score', y='CGPA', data=df, fit_reg=False, hue='University Rating')
```

```python
uni_type_colors = ['#78C850',
                   '#C03028',
                   '#A040A0',
                   '#B8A038',
                   '#7038F8']
```

We use swarm plot to showw the correlation between University Rating and GRE Score:

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="University Rating", y='GRE Score', data=df,
         palette=uni_type_colors)
```

```python
sns.histplot(data =df, x = 'GRE Score', kde=True)
```

```python
sns.histplot(data =df, x = 'Chance of Admit ')
```

```python
df1 = df.groupby('University Rating')['Research'].apply(lambda x: (x==1).sum()).reset_index(name = 'research_count')
```

```python
sns.barplot(x='University Rating', y = 'research_count', data=df1)
```

```python
sns.kdeplot(x='GRE Score', y='Chance of Admit ', data=df)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(x='GRE Score', y='Chance of Admit ', data=df, cmap="Blues", fill=True, shade=True,n_levels=15)
```
